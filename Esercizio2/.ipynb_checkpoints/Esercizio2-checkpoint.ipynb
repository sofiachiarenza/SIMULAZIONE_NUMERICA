# Esercizio 2.1 - <span style="color:deeppink"> Importance sampling </span>

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import norm , cauchy
import matplotlib.mlab as mlab
from scipy.optimize import curve_fit
```

```python
sum_prog , error_prog , sum_prog_IS , error_prog_IS , sum_prog_IS_2 , error_prog_IS_2 = np.loadtxt( "Esercizio2.1/data.csv" , dtype = float , delimiter = "," , usecols=(0,1,2,3,4,5) , unpack='true')
```

```python
fig, [ax0 , ax1 , ax2 ]   = plt.subplots(1, 3, figsize=(22 , 8))
ax0.errorbar( np.arange(len(sum_prog)) , sum_prog , yerr=error_prog , color="navy" , label="Metodo della media" )
ax1.errorbar( np.arange(len(sum_prog)), sum_prog_IS , yerr=error_prog_IS , color="purple" , label="Importance Sampling")
ax2.errorbar(np.arange(len(sum_prog)), sum_prog_IS_2 , yerr=error_prog_IS_2 , color="orange" , label="Different importance Sampling")

for i in range(3):
    globals()['ax'+str(i)].set_title(r'Extimations for $\int_0 ^1 \frac{\pi}{2} cos(\frac{\pi x}{2})$' , fontsize = 16)
    globals()['ax'+str(i)].set_ylabel(r'I')
    globals()['ax'+str(i)].set_xlabel('Blocks')
    globals()['ax'+str(i)].legend()
    globals()['ax'+str(i)].grid(True)
```

I tre grafici mostrano la stima dell'integrale all'aumentare del numero di tiri. Le stime riportate nei grafici e il relativo errore sono state calcolate con il metodo delle medie a blocchi.
Nel primo grafico l'integrazione è stata eseguita utilizzando il metodo della media, una tecnica Monte Carlo standard per il calcolo di integrali.
Nel secondo grafico invece, ho provato ad applicare il metodo dell'importance sampling usando una distribuzione $d(x) = 2 cos^2(\frac{\pi}{2}x )$. Il risultato non è ottimale, poichè è evidente dai grafici che non è stata ottenuta una significativa riduzione della varianza come sperato.
Il terzo grafico riporta quindi un terzo tentativo di importance sampling, questa volta usando una distribuzione $d(x) = \frac{3}{2}(1 - x^2)$. La varianza ha una riduzione molto più importante in questo caso.

```python
plt.errorbar( np.arange(len(sum_prog)) , sum_prog , yerr=error_prog , color="navy", label="Metodo della media" )
plt.errorbar( np.arange(len(sum_prog)) , sum_prog_IS , yerr=error_prog_IS , color="purple" , label="Importance Sampling")
plt.errorbar( np.arange(len(sum_prog)) , sum_prog_IS_2 , yerr=error_prog_IS_2 , color="orange", label="Differente Importance Sampling")
plt.xlabel('Blocks')
plt.ylabel(r'I')
plt.title(r'Extimations for $\int_0 ^1 \frac{\pi}{2} cos(\frac{\pi x}{2})$ computed with different methods' , fontsize = 16)
plt.grid(True)
plt.legend(loc=(1.1 , 0.7))
plt.show()
```

In questo grafico ho riportato gli stessi tre grafici precedenti, ma sovrapponendoli in modo che le differenze siano meglio apprezzabili. Come anticipato non si vede una differenza significatriva tra il metodo della media e l'importance sampling con una distribuzione di tipo $cos^2(x)$. Nel terzo caso invece è molto netto il miglioramento sia della stima dell'integrale, molto meno oscillante e più vicina al valore vero, sia della varianza, che è notevolmente ridotta.

```python
plt.plot(np.arange(len(sum_prog)) ,error_prog , color="navy", label="Metodo della media" )
plt.plot(np.arange(len(sum_prog)) ,error_prog_IS , color="purple" , label=r"Importance sampling $d(x)=2 cos^2(\frac{\pi}{2}x)$")
plt.plot(np.arange(len(sum_prog)) ,error_prog_IS_2 , color="orange", label=r"Importance sampling $d(x)=\frac{3}{2}(1-x^2)$")
plt.xlabel('Blocks')
plt.ylabel(r'$\sigma ^2$')
plt.title("Variance as a function of blocks")
plt.grid(True)
plt.legend(loc=(1.1 , 0.7))
```

Ho riportato in questo grafico le varianze nei 100 blocchi in cui è stato suddiviso il totale di $10^5$ punti generati per la stima dell'integrale, in modo da rendere ancora più evidente quanto la distribuzione $d(x)=\frac{3}{2}(1-x^2)$ sia efficacie ai fini di ridurre la varianza. Per quanto riguarda invece $d(x)=2 cos^2(\frac{\pi}{2}x)$, all'aumentare del numero di tiri la varianza tende a coincidere con quella del metodo della media, risultato non auspicabile dato che l'obiettivo dell'importance sampling è la riduzione della varianza.

# Esercizio 2.2 - Random Walks

```python
dist , errors , dist2 , errors2 = np.loadtxt( "Esercizio2.2/data.csv" , dtype = float , delimiter = "," , usecols=(0,1,2,3) , unpack='true')
```

```python
fig, (ax0 , ax1)   = plt.subplots(1, 2, figsize=(20 , 8))

ax0.errorbar( np.arange(len(dist)) , dist , yerr= errors ,color="mediumvioletred", label="Computed" )
ax1.errorbar( np.arange(len(dist2)) , dist2 , yerr= errors2 ,color="mediumvioletred", label="Computed" )

ax0.set_xlabel('Steps')
ax0.set_ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$' , fontsize = 15)
ax0.set_title(r"Random walk in a cubic lattice" , fontsize = 18)

ax1.set_xlabel('Steps')
ax1.set_ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$' , fontsize = 15)
ax1.set_title(r"Random walk in a 3D space" , fontsize = 18)

ax0.grid(True)
ax1.grid(True)

#fitting the curve 

def f(x,a):  # defining the fitting function
    return a * np.sqrt(x) 

x = np.linspace(0, 100 , 100) ;

#fitting per il random walk nel reticolo cubico
p_opt, p_cov = curve_fit(f, x, dist)
y_fit = f(x,p_opt[0])
ax0.plot(x, y_fit , color="tomato" , label = r"Fit: $f(x)= $"+str(np.around(p_opt[0],3))+"$ \sqrt{x}$") 

#fitting per il random walk nel continuo
p_opt2, p_cov2 = curve_fit(f, x, dist2)
y_fit2 = f(x,p_opt2[0])
ax1.plot(x, y_fit2 , color="tomato" , label = r"Fit: $f(x)= $"+str(np.around(p_opt2[0],3))+"$ \sqrt{x}$") 

ax0.legend()
ax1.legend()

print("optimized parameters [a , a'] =")
print("[" , p_opt[0] ,", " ,  p_opt2[0] , "]")
print("parameters uncertainty =")
print("[" , np.sqrt(np.diagonal(p_cov)) , "," , np.sqrt(np.diagonal(p_cov2)) , "]" )
```

I grafici riportano la distanza quadratica media del random walk dall'origine, in funzione del numero di passi compiuti. Nel primo caso il random walk è vincolato ad un reticolo cubico, mentre nel secondo grafico il random walk si trova nel continuo. In entrambi i casi la lunghezza del passo è la stessa.
E' noto che la distribuzione di probabilità della fine del random walk rispetta l'equazione di diffusione $$ \frac{\partial \rho}{\partial t} = D \nabla ^2 \rho $$ dove D assume in questo caso il valore $D = \frac{a^2}{2 \tau}$, con a lunghezza del passo, e $\tau$ tempo che intercorre tra gli step (in questa simulazione quindi D = $1/2$).  L'equazione di diffusione si può risolvere semplicemente imponendo una condizione iniziale, nel nostro caso la partenza dall'origine, e passando in spazio di Fourier. Quello che si trova è che la distribuzione di probabilità della posizione finale del cammino è una gaussiana centrata nell'origine, e con deviazione standard $\sigma = \sqrt{2DN\tau} = \sqrt{N} $, che è ciò che troviamo dalla simulazione.
