# Esercizio 4.1 - <span style="color:deeppink"> Equilibration of a solid system </span>

```python
e_kin  = np.loadtxt( "MolecularDynamics_NVE/equilibration/output_ekin.dat" , dtype = float , delimiter = "," , usecols=(0) , unpack='true')
e_pot  = np.loadtxt( "MolecularDynamics_NVE/equilibration/output_epot.dat" , dtype = float , delimiter = "," , usecols=(0) , unpack='true')
e_tot  = np.loadtxt( "MolecularDynamics_NVE/equilibration/output_etot.dat" , dtype = float , delimiter = "," , usecols=(0) , unpack='true')
temp  = np.loadtxt( "MolecularDynamics_NVE/equilibration/output_temp.dat" , dtype = float , delimiter = "," , usecols=(0) , unpack='true')

#SI UNITS
e_kin *= epsilon 
e_pot *= epsilon
e_tot *= epsilon
temp *= 120 
```

```python
fig, ax = plt.subplots(1,1 , figsize=(13,8))
ax.plot(np.arange(len(temp)) , temp , color="#ECA5E8" , linewidth=3 , label = "Temperature")
ax.hlines(0.8*120 , 0 , 500 , color = "Purple" , linewidth = 3 , label ="Expected temperature")
ax.set_title("Temperature in the equilibration of a solid system" , fontsize=25)
ax.set_xlabel('Time steps')
ax.set_ylabel('T (K)')

ax.grid(True)
ax.legend(loc="best")
```

```python
fig, ax = plt.subplots(1,1 , figsize=(13,8))
ax.plot(np.arange(len(e_kin)) , e_kin , color = "#F3C16F", linewidth=3, label = "Kinetic energy")
ax.plot(np.arange(len(e_kin)) , e_pot , color = "#CCCCFF" , linewidth=3 , label = "Potential energy")
ax.plot(np.arange(len(e_kin)) , e_tot , color = "#A5EDB5" , linewidth=3 , label = "Total energy")

ax.set_title("Energy in the equilibration of a solid system" , fontsize=25)
ax.set_xlabel('Time steps')
ax.set_ylabel('Energy (J)')

ax.grid(True)
ax.legend(loc="best")
```

    I grafici mostrano la fase di equilibrazione di un sistema solido di 108 particelle. Le equazioni del moto sono state integrate con l'algoritmo di Verlet. In ogni simulazione sono stati eseguiti 500 passi di integrazione dell'algoritmo, ma la misura di energia e temperatura è stata eseguita ogni 10 passi per evitare un'eccessiva correlazione. È necessario includere una fase di equilibrazione, perchè l'algoritmo di Verlet non è esatto, ci vuole quindi del tempo prima che inizi a simulare alla temperatura desiderata. Sono riportati i grafici per un sistema solido, il quale si equilibra dopo 5 ripartenze a distanza di 500 step ciascuna.
    Simili prove sono state fatte per sistemi liquidi e gassosi, in questi casi ho osservato un tempo di equilibrazione maggiore (10 ripartenze per il liquido, e 20 per i gas). Questo accade perchè sistemi liquidi e gassosi sono più disordinati di quello solido, e l'algoritmo di Verlet fatica maggiormente a stabilizzarsi.

```python
eksolid , errorksol = np.loadtxt( "MolecularDynamics_NVE/solid/ave_ekin.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')
epotsol , errorpsol = np.loadtxt( "MolecularDynamics_NVE/solid/ave_epot.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')
etotsol , errortsol = np.loadtxt( "MolecularDynamics_NVE/solid/ave_etot.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')
tempsol , errortempsol = np.loadtxt( "MolecularDynamics_NVE/solid/ave_temp.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')
pressol , errorpressol = np.loadtxt( "MolecularDynamics_NVE/solid/ave_pres.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')

#SI UNITS
eksolid *= epsilon 
epotsol *= epsilon
etotsol *= epsilon
tempsol *= 120 
errorksol *= epsilon
errorpsol *= epsilon
errortsol *= epsilon
errortempsol *= 120
pressol *= epsilon/sigma**3
errorpressol *= epsilon/sigma**3
```

La fase gassosa dell'argon è stata simulata a T = 144 K, $\rho = 1.96 * 10^{-30}$ e con cut-off $r = 1.7 * 10^{-9}$. Il sistema è stato equilibrato eseguendo 20 ripartenze separate da 500 step temporali, la simulazione in sè invece fa 20000 step, che sono stati poi suddivisi in 50 blocchi. L'andamento delle energie è buono, in particolare l'energia totale rimane costante quindi è conservata, come dovrebbe essere. La temperatura tuttavia non risulta perfettamente compatibile con quella di simulazione.

Il fatto che nessuna temperatura risulti del tutto compatibile è dovuto alla scarsa accuratezza dell'algoritmo di Verlet, che è un algoritmo deterministico che integra in modo approssimato le equazioni del moto. Ogni step temporale dell'algoritmo porta con sè un errore di discretizzazione, errore che si propaga nel tempo di simulazione. Nel nostro caso una simulazione è di 20000 step , ciascuno di 0.0005 secondi in unità di LJ, ossia $1.07 * 10^{-15}$ secondi, per un totale di $2.14 * 10^{-11}$ secondi di simulazione. L'errore che si propaga su 20000 step è abbastanza piccolo perchè l'energia appaia conservata, mentre ha un peso maggiore sulla temperatura, che evidentemente non rimane costante durante la simulazione.

### Conversion from Lennard-Jones units to SI units

Le unità di Lennard Jones sono convenienti perchè adimensionali, il loro ordine di grandezza inoltre evita problemi di underflow, che possono verificarsi soprattutto nello stimare l'energia, che se misurata in joule ha ordine di grandezza $10^{-21}$. Nel caso dell'argon, l'unità di lunghezza $\sigma$ è 0.34 nm , l'unità di energia $ \frac{\epsilon} { k_{b}}$ è 120 K e l'unità di massa è $m = 39,948 $ amu. 

Per riconvertire da unità di Lennard Jones ad unità di misura del SI, bisogna: 
1. Moltiplicare le lunghezze per $\sigma$ : $ r = \sigma r^* $
2. Moltiplicare le temperature per $ \frac{\epsilon} { k_{b}} $ : $ T = T^* \frac{\epsilon} { k_{b}} $
3. Moltiplicare le energie per $\epsilon$: $ E = E^* \epsilon $
4. Moltiplicare il tempo per $  \sqrt ( \frac {m \sigma^2}{ \epsilon} ) $
